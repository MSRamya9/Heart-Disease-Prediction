```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
heart_data.isnull().sum()
```

```python
heart_data.shape
```

```python
heart_data['target'].value_counts()
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
model1=SVC()
```

```python
model2=MultinomialNB()
```

```python
model3 = LogisticRegression()
```

```python
model4=RandomForestClassifier()
```

**SVM**

```python
model1.fit(X_train, Y_train)
```

```python
X_train_prediction = model1.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data (SVM) : ', training_data_accuracy*100)
```

```python
X_test_prediction = model1.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data (SVM) : ', test_data_accuracy*100)
```

```python
Y_combined1 = np.concatenate([Y_train, Y_test])
Y_pred_combined1 = np.concatenate([X_train_prediction, X_test_prediction])
# Calculate the overall accuracy
overall_accuracy = accuracy_score(Y_combined1, Y_pred_combined1)
print('Overall Accuracy (SVM):', overall_accuracy*100)
# Calculate the confusion matrix
confusion_matrix_svm = confusion_matrix(Y_combined1, Y_pred_combined1)
print('Confusion Matrix (SVM):')
print(confusion_matrix_svm)
```

```python
x=training_data_accuracy*100
y=test_data_accuracy*100
plt.xlim(1,100)
plt.ylim(1,100)
plt.grid()
plt.plot(x, y, marker="o", markersize=10, markeredgecolor="black",
markerfacecolor="purple")
plt.show()
```
